### Importing Essential Packages

### Reading the data and fetching the relevant information from the data

```python
data = pd.read_csv("../input_data/AdmissionDataset/data.csv")

#Dropping the Serial Number field of the given input table as it is not so relevant information for any prediction.
data.drop(data.columns[[0]], axis=1, inplace=True)
```

### Normalising the attributes so that the contours of the cost function will not form skewed elliptical shape and hence converge more quickly. 

```python
#Normalizing all the attributes except for Research and Chance of Admit (as they are already in range) . 
def normalize(data):
    for feature_name in data.columns[:-2]:
        mean_value = data[feature_name].mean()
        max_value = data[feature_name].max()
        min_value = data[feature_name].min()
        data[feature_name] = (data[feature_name] - mean_value) / (max_value - min_value)
    return data    
```

### After normalizing the dataset , split the data into train and validation subsets into 80:20 ratio.

```python
data = normalize(data)

#For splitting the initial 80% and 20% dataset for train and validation

train,validate = np.split(data,[int(.8*len(data))])
train = train.reset_index(drop = True)
validate = validate.reset_index(drop = True)

# For Random Splitting

# data=data.sample(frac=1).reset_index(drop=True)
# train, validate = np.split(data, [int(.8*len(data))])

# train
```

### Creating a dataset X that contains the first column as all 1's and which doesn't contain the final label 'Chance of Admit'(Input data)

```python
def find_X_and_Y():
    #Adding a column 'ones' initialized with value 1 for all rows.
    X = train.iloc[:,0:7]
    ones = np.ones([X.shape[0],1])
    X = np.concatenate((ones,X),axis=1)
    Y = np.array([train.iloc[:, -1].values]).T
    return X,Y
```

```python
X,Y = find_X_and_Y()
print (X)
```

### Initialising the values of alpha and number of iterations.

```python
alpha = 0.001
num_iter = 100
theta = np.zeros([1,X.shape[1]])
theta
```

```python
def mse(X, Y, theta, L):
    errors = np.dot(X , theta.T) - Y
    square_errors = np.power(errors, 2) + L * np.sum(np.square(theta), axis=0)
    return errors, square_errors, (np.sum(square_errors) / (2 * len(X)))
```

```python
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))
```

```python
def gradient_descent(X, Y, theta, nIterations, alpha, L):
    cost_list = []
    
    
    for i in range(nIterations):
#         print("X", X.shape)
#         print("Y", Y.shape)
#         print(" theta ",theta.shape)
        temp_X = X[:,1:]
        temp_theta0 = theta[0]
        temp_theta = theta[:,1:]
#         print("temp X", temp_X.shape)
#         print("temp Y", Y.shape)
#         print(" temp_theta ",temp_theta.shape)
        temp_theta0 = temp_theta0 - (alpha/len(X)) * (np.array([np.sum((temp_theta0 * (X @ theta.T - Y)), axis = 0)]))
        temp_theta = temp_theta*(1 - L*(alpha/len(X))) - (alpha/len(X)) * (np.array([np.sum((temp_X * (X @ theta.T - Y)), axis = 0)]))
        theta[0] = temp_theta0
        theta[:,1:] = temp_theta
        
        current_cost = mse(X, Y, theta, L)[2]
        cost_list.append(current_cost)
    return theta, cost_list
```

```python
lam_arr = np.arange(-20,10,0.5)
c_list = []
for lam in lam_arr:
    theta,cost_list = gradient_descent(X,Y,theta,num_iter,alpha,lam)
    print("lam : ",cost_list[-1])
    print("-----------------------------------------------\n")
    c_list.append(cost_list[-1])
print(theta)
# finalCost = computeCost(X,Y,theta)
# print(finalCost)
```
