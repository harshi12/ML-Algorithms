```python
def Hypothesis_H(Theta,Row):
    Hx = Theta[0]
    for i in range(1,len(Row)):
        Hx += Theta[i]*Row[i-1]
    return Hx
```

```python
def Cost_Fun(Theta_MSE,Theta_MAE,Theta_MAPE,Training_Data):
    Cost_MSE = 0
    Cost_MAE = 0
    Cost_MAPE = 0
    for Index_no, Training_Row in Training_Data.iterrows():
        Cost_MSE += (Hypothesis_H(Theta_MSE,Training_Row) - Training_Row[Class])**2
        Cost_MAE += abs(Hypothesis_H(Theta_MAE,Training_Row) - Training_Row[Class])
        Cost_MAPE += abs((Training_Row[Class] - Hypothesis_H(Theta_MAPE,Training_Row))/Training_Row[Class])
    Cost_MSE = Cost_MSE/(2*len(Training_Data))
    Cost_MAE = Cost_MAE/len(Training_Data)
    Cost_MAPE = (Cost_MAPE/len(Training_Data))
    return Cost_MSE, Cost_MAE, Cost_MAPE
```

```python
def List_Of_Error(Theta_MSE,Theta_MAE,Theta_MAPE,Training_Data):
    Error_MSE = []
    Error_MAE = []
    Error_MAPE = []
    for Index_no, Training_Row in Training_Data.iterrows():
        Error_MSE.append(Hypothesis_H(Theta_MSE,Training_Row) - Training_Row[Class])
        
        if(Hypothesis_H(Theta_MAE,Training_Row) - Training_Row[Class] > 0):
            Error_MAE.append(1)
        else:
            Error_MAE.append(-1)
            
        if(((Training_Row[Class] - Hypothesis_H(Theta_MAPE,Training_Row))/Training_Row[Class]) > 0):
            Error_MAPE.append((-1)/Training_Row[Class])
        else:
            Error_MAPE.append((1)/Training_Row[Class])
            
    return Error_MSE,Error_MAE,Error_MAPE
```

```python
def Calculate_For_Error_MSE(Error_List,Jth_Column):
    Ans = 0
    for i in range (0,len(Error_List)):
        Ans += Error_List[i] * Jth_Column[i]
    return Alpha * (Ans/len(Error_List))
```

```python
def Calculate_For_Error_MAE(Error_List,Jth_Column):
    Ans = 0
    for i in range (0,len(Error_List)):
        Ans += Error_List[i] * Jth_Column[i]
    return Alpha * (Ans/len(Error_List))
```

```python
def Calculate_For_Error_MAPE(Error_List,Jth_Column):
    Ans = 0
    for i in range (0,len(Error_List)):
        Ans += Error_List[i] * Jth_Column[i]
    return Alpha * (Ans/len(Error_List))
```

```python
def Gradiant_Descent(Theta_MSE,Theta_MAE,Theta_MAPE,Training_Data,Error_MSE, Error_MAE, Error_MAPE):
    for j in range(0,len(Theta_MSE)):
        if (j == 0):
            Theta_MSE[j] -= Calculate_For_Error_MSE(Error_MSE,[1]*len(Training_Data))
            Theta_MAE[j] -= Calculate_For_Error_MAE(Error_MAE,[1]*len(Training_Data))
            Theta_MAPE[j] -= Calculate_For_Error_MAPE(Error_MAPE,[1]*len(Training_Data))
        else:
            Theta_MSE[j] -= Calculate_For_Error_MSE(Error_MSE,Training_Data.iloc[:,j-1])
            Theta_MAE[j] -= Calculate_For_Error_MAE(Error_MAE,Training_Data.iloc[:,j-1])
            Theta_MAPE[j] -= Calculate_For_Error_MAPE(Error_MAPE,Training_Data.iloc[:,j-1])
    return Theta_MSE,Theta_MAE,Theta_MAPE 
        
```

```python
def Find_Mean_Square_Error(Theta,Validation_Data):
    Mean_Square_Error = 0
    for Index_no, Validation_Row in Validation_Data.iterrows():
        Mean_Square_Error += (Validation_Row[Class] - Hypothesis_H(Theta,Validation_Row))**2
    return Mean_Square_Error/len(Validation_Data)
```

```python
def Find_Mean_Absolute_Error(Theta,Validation_Data):
    Mean_Absolute_Error = 0
    for Index_no, Validation_Row in Validation_Data.iterrows():
        Mean_Absolute_Error += abs(Validation_Row[Class] - Hypothesis_H(Theta,Validation_Row))
    return Mean_Absolute_Error/len(Validation_Data)
```

```python
def Find_Mean_Absolute_Percentage_Error(Theta,Validation_Data):
    Mean_Absolute_Percentage_Error = 0
    for Index_no, Validation_Row in Validation_Data.iterrows():
        Mean_Absolute_Percentage_Error += (abs(Validation_Row[Class] - Hypothesis_H(Theta,Validation_Row))/Validation_Row[Class])

    return ((Mean_Absolute_Percentage_Error)/len(Validation_Data))*100
```
