```python
def cost_function(theta,df):
    m =len(df)
    error = hypothesis(theta,df)
    J = 0
    for i in range(m):
        J += np.square(error[i])
    avg = J/(2*m)
    return avg,error
    
```

```python
def gradient_descent(error,df,theta,learning_rate):
    m = len(df)
    for j in range(7):
        i = 0
        sum_error = 0
        for index, row in df.iterrows():
            sum_error += error[i]*row[j]
            i += 1
        theta[j] = theta[j] - (learning_rate/m)*sum_error
    theta[7] = theta[7] - (learning_rate/m)*(sum(error))
    return theta
```

```python
def hypothesis(theta,df):
    #     t = np.array(theta)
    error = []
    for index, row in df.iterrows():
        hypo = 0
        y = row[7]
        for i in range(len(row)-1):
            hypo += theta[i] * row[i]
        hypo += theta[7]
        error.append(hypo - y)
    return error
```

```python
import pandas as pd
import numpy as np
import math
import operator
from numpy import log2 as log
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
eps = np.finfo(float).eps
from matplotlib import pyplot as plt

df = pd.read_csv("./../input_data/AdmissionDataset/data.csv")
# df = df.sample(frac=1).reset_index(drop=True)
df = df.drop('Serial No.', axis=1)

cols = list(df)
for i in cols[:-1]:
    df[i] = (df[i] - df[i].min())/(df[i].max() - df[i].min()) 
```

```python



training, validate = np.split(df,[int(0.8*len(df))])
length = len(list(training))

theta = [0] * (length)
m = len(training)
# normalizing the dataset-------


# training = (training - training.mean())/training.std()

# print(" DATA AFTER NORMALIZATION")
# print(training)
no_of_iterations = 1000
learning_rate = 0.01
cost = []
iteration = []

for i in range(no_of_iterations):
#     print(theta)
    c,error = cost_function(theta,training)
    cost.append(c)
    theta = gradient_descent(error,training,theta,learning_rate)
    iteration.append(i)
    print(i, c)  


```

```python
def predict(theta, featureRow):
    n = len(theta)
    y = 0
    for i in range(0, n - 1):
        y += theta[i]*featureRow[i]
    return y + theta[n - 1]
```

```python
def valdidation(validate,theta):
    predicted = []
    actual = []
    for index, row in validate.iterrows():
        pred = predict(theta, row)
        ac = row[-1]
        predicted.append(pred)
        actual.append(ac)
        print(pred)
        print(ac)
        print("-----------")
    return predicted , actual

```

```python
predicted , actual = valdidation(validate,theta)
```

```python
#Mean square error
def MSE(predicted , actual):
    return np.mean((np.array(actual) - np.array(predicted))**2)
```

```python
#Mean absolute error
def MAE(predicted , actual):
    return np.mean(np.abs((np.array(actual) - np.array(predicted))))
```

```python
# Mean absolute percentage error function
def MAPE(predicted , actual):
    return np.mean(np.abs((np.array(actual) - np.array(predicted))/ np.array(actual))) * 100
```
