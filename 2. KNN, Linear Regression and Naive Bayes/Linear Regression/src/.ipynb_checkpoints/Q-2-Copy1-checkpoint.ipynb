```python
def mean(col_values):
    return (sum(col_values) / float(len(col_values)))
```

```python
from math import sqrt

def stdv(col_values):
    avg = mean(col_values)
    var = sum([pow(i - avg, 2) for i in col_values]) / float(len(col_values) -1)
    return sqrt(var)
```

```python
from math import exp

def gaussian_probab(val, mean, stdv):
    exponent = math.exp(-(math.pow(val-mean,2)/(2*math.pow(stdv,2))))
    return (1 / (math.sqrt(2*math.pi) * stdv)) * exponent
```

```python
import pandas as pd
import numpy as np
import math
import operator
from numpy import log2 as log
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
eps = np.finfo(float).eps

df = pd.read_csv("../Input Data/LoanDataset/data.csv", header = None)
df.columns = ['id', 'age', 'experience', 'income', 'zip', 'family_size', 'spending', 'education','mortgage', 'output', 'security_account','certi_deposit','internet','credit_card']
df = df.sample(frac=1).reset_index(drop=True)
train_data, val_data = np.split(df,[int(0.8*len(df))])

data= ['id', 'age', 'experience', 'income', 'zip', 'family_size', 'spending', 'education','mortgage', 'output', 'security_account','certi_deposit','internet','credit_card']
num_data = ['age', 'experience', 'income', 'zip', 'family_size', 'spending', 'education','mortgage']

# to store mean and standard deviation of numerical data
mean_dic_0 = {}
stdv_dic_0 = {}
mean_dic_1 = {}
stdv_dic_1 = {}

for i in data:
    if i != 'output' and i != 'id':
        mean_dic_0[i] = mean(df[df['output'] == 0][i])
        stdv_dic_0[i] = stdv(df[df['output'] == 0][i])
        mean_dic_1[i] = mean(df[df['output'] == 1][i])
        stdv_dic_1[i] = stdv(df[df['output'] == 1][i])
    
print(mean_dic_0)
print(stdv_dic_0)
print(mean_dic_1)
print(stdv_dic_1)
```

```python
freq = df['age'].value_counts()
freq_list = []
values = df['age'].unique()
values.sort()
probab = []
for i in values:
    freq_list.append(freq[i])
    probab.append(gaussian_probab(i, mean(df['age']) , stdv(df['age'])))

print(probab)
plt.figure(num=None, figsize=(8, 6), dpi=200, facecolor='w', edgecolor='k')
plt.ylabel("Age")
plt.xlabel("Frequency")
plt.title('Finding distribution')
plt.grid(True)
# plt.plot( values,freq_list, marker='o', color='green', label = 'Error rate')
plt.plot(  values,probab, marker='+', color='blue', label = 'Error rate')
plt.show()
```

```python
cat_data = ['security_account_0', 'security_account_1', 'certi_deposit_0', 'certi_deposit_1', 'internet_0', 'internet_1', 'credit_card_0', 'credit_card_1', 'total']
cat_probab = []
for i in range(2): #third sublist is for the totals so is the last entry of each sublist
    temp = [0 for i in range(len(cat_data))]
    cat_probab.append(temp)

#populate the probab matrix for categorical data
for k in range(2):
    sum_hor = 0
    for i in range(len(cat_data)):
        if cat_data[i] == 'total':
            cat_probab[k][i] = sum_hor
        else:
            label = cat_data[i][:-2] #get the name of the attribute
            val = int(cat_data[i][-1]) # will give 0 or 1
            temp = df[df['output'] == k][label].value_counts()
            cat_probab[k][i] = temp[val]
            sum_hor += temp[val]


print(cat_probab)
# cat_data.index('total')
```

```python
predicted = []
tot_output = df['output'].value_counts()

for index, row in val_data.iterrows():
#     print(row)
    probab_0 = 1
    probab_1 = 1
    num_0 = [] # for p(A/ output = 0)
    num_1 = [] # for p(A/ output = 1)
    for i in range(row.shape[0]):
        label = data[i]
        if label == 'id' or label == 'output':
            continue
        
        elif label not in num_data:
            if row[i] == 0:
                label += '_0'
            else:
                label += '_1'
#             print(label)
            var1 = cat_probab[0][cat_data.index(label)] + cat_probab[1][cat_data.index(label)]
            var2 = cat_probab[0][-1] + cat_probab[1][-1]
            num_0.append(float((cat_probab[0][cat_data.index(label)] * pow(var2, 2))))
            num_1.append(float((cat_probab[1][cat_data.index(label)] * pow(var2,2))))
        
        else:
            op_0 = tot_output[0] / (eps + len(df))
            op_1 = tot_output[1] / (eps + len(df))
            temp = gaussian_probab(row[i], mean_dic_0[label], stdv_dic_0[label])
            num_0.append(temp * op_0)
            temp = gaussian_probab(row[i], mean_dic_1[label], stdv_dic_1[label])
            num_1.append(temp * op_1)

    num_0.append(cat_probab[0][-1] / float(eps + cat_probab[0][-1] + cat_probab[1][-1]))
    num_1.append(cat_probab[1][-1] / float(eps + cat_probab[0][-1] + cat_probab[1][-1]))
    
    for i in num_0:
        if i != 0.0:
            probab_0 *= i
    for i in num_1:
        if i != 0.0:
            probab_1 *= i
    if probab_0 > probab_1:
        predicted.append(0)
    else:
        predicted.append(1)
#     print (probab_0)
#     print (probab_1)
#     print("output:  ",row[-5])
#     break
    
# print(len(val_data['output']))
# print(len(predicted))
    
print(accuracy_score(predicted, val_data['output']))
```
