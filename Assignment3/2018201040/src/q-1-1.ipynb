```python
import numpy as np
import pandas as pd
import math as mt
import sys
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
# from sklearn.preprocessing import MinMaxScaler

eps = np.finfo(float).eps
```

```python
dataSet = pd.read_csv("./../input_data/data 2.csv")
Class = "xAttack"
columns = list(dataSet.columns)
X = dataSet[columns[:-1]]

```

```python
X.head(10)
```

```python
Y = dataSet[[Class]]
```

```python
Y.head(10)
```

```python
x_std = StandardScaler().fit_transform(X)
```

```python
x_std
```

```python
features = x_std.T 
covariance_matrix = np.cov(features)
covariance_matrix
```

```python
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
```

```python
eig_vecs
```

```python
eig_vals
```

```python
eig_dic = {}
for i in range(0,len(eig_vals)):
    eig_dic[eig_vals[i]] = eig_vecs[:,i]
cum_sum = 0
select_eig_vecs = []
attr_name = []
c = 1
total = sum(eig_vals)
for key in reversed(sorted(eig_dic.keys())):
    cum_sum += abs(key)
    correct_res = cum_sum/total
    select_eig_vecs.append(eig_dic[key])
    attr_name.append("Col-"+ str(c))
    c += 1
    if correct_res > 0.90:
        break
print(c) 
correct_res
```

```python
projected_X = x_std.dot(np.array(select_eig_vecs).T)
projected_X
```

```python
result = pd.DataFrame(projected_X, columns = attr_name)
result[Class] = Y
```

```python
result.to_csv("./../output_data/q-1-1_output.csv",index = False)
```
