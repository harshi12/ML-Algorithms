### PART 1 : Bandwidth Optimization

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
from sklearn.neighbors import KernelDensity
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
```

```python
# load the data
digits = load_digits()
```

```python
df_matrix = digits.data
df_matrix
```

```python
pca_for_31 = PCA(n_components=31, whiten=False)
recontructed_data_31 = pca_for_31.fit_transform(df_matrix)
pca_for_35= PCA(n_components=35, whiten=False)
recontructed_data_35 = pca_for_35.fit_transform(df_matrix)
pca_for_41 = PCA(n_components=41, whiten=False)
recontructed_data_41 = pca_for_41.fit_transform(df_matrix)
```

```python
params = {'bandwidth': np.logspace(-1, 1)}
grid = GridSearchCV(KernelDensity(), params, iid=True,cv=5)
grid.fit(recontructed_data_31)
kde31 = grid.best_estimator_
print(kde31)
print("best bandwidth: {0}".format(grid.best_estimator_.bandwidth))
```

```python
params = {'bandwidth': np.logspace(-1, 1)}
grid = GridSearchCV(KernelDensity(), params, iid=True,cv=5)
grid.fit(recontructed_data_35)
kde_for_35 = grid.best_estimator_
print(kde_for_35)
print("best bandwidth: {0}".format(grid.best_estimator_.bandwidth))
```

```python
params = {'bandwidth': np.logspace(-1, 1)}
grid = GridSearchCV(KernelDensity(), params, iid=True,cv=5)
grid.fit(recontructed_data_41)
kde_for_41 = grid.best_estimator_
print(kde_for_41)
print("best bandwidth: {0}".format(grid.best_estimator_.bandwidth))
```

### PART 2: GMM (finding number of components)

```python

from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3).fit(df_matrix)
labels = gmm.predict(df_matrix)
plt.scatter(df_matrix[:, 0], df_matrix[:, 1], c=labels, s=40, cmap='viridis');

```

```python
n_components = np.arange(1, 21)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(df_matrix) 
          for n in n_components]

plt.plot(n_components, [m.bic(df_matrix) for m in models], label='BIC')
plt.plot(n_components, [m.aic(df_matrix) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
```

```python
def plot_digits(data):
    fig, ax = plt.subplots(8, 6, figsize=(10, 10))
    for i, axi in enumerate(ax.flat):
        im = axi.imshow(data[i].reshape(8, 8), cmap = 'binary')
        im.set_clim(0, 16)

# def plot_digits(data):
#     fig, ax = plt.subplots(8, 6, figsize=(10, 10),
#                            subplot_kw=dict(xticks=[], yticks=[]))
#     fig.subplots_adjust(hspace=0.05, wspace=0.05)
#     for i, axi in enumerate(ax.flat):
#         im = axi.imshow(data[i].reshape(8, 8), cmap='binary')
#         im.set_clim(0, 16)
# plot_digits(df_matrix)

```

### We will start with an invertible dimensionality reduction algorithm on the data. Here we will use a straightforward PCA, asking it to preserve 99% of the variance in the projected data:

```python
from sklearn.decomposition import PCA
# pca = PCA(0.99, whiten=True)
data = pca_for_41.fit_transform(df_matrix)
data.shape
```

### The result is 41 dimensions, a reduction of nearly 1/3 with almost no information loss. Given this projected data, let's use the BIC (Bayesian information criteria) to get a gauge for the number of GMM components we should use:

```python
n_components = np.arange(50, 210, 10)
models = [GaussianMixture(n, covariance_type='full', random_state=0)
          for n in n_components]
bics = [model.fit(data).bic(data) for model in models]
plt.plot(n_components, bics);

```

```python
def reconstructkde(kernel,pca):
    data_new = kernel.sample(48, random_state=0)
    
    digits_new = pca.inverse_transform(data_new)
    
    plot_digits(digits_new)
```

```python
rec_data = reconstructkde(kde_for_41, pca_for_41)
```

### From the above graph it appears that around 85 components minimizes the BIC; we will use this model.

```python
gmm = GaussianMixture(85, covariance_type='full', random_state=0)
gmm.fit(recontructed_data_41)
print(gmm.converged_)
```

### PART 3 (Inverse transform of PCA)

### Now we can draw samples of 48 new points within this 41-dimensional projected space, using the GMM as a generative model:

```python
data_new_x,y = gmm.sample(48)
data_new_x.shape
```

```python
digits_new = pca_for_41.inverse_transform(data_new_x)
plot_digits(digits_new)
```
