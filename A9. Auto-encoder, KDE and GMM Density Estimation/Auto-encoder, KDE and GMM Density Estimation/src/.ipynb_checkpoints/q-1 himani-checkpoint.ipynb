```python
def standardize_dataset(ds,label):
    for col in ds.columns:
        if col==label:
            continue
        else:
            mean = ds[col].mean()
            std = ds[col].std()
            ds[col] = (ds[col]-mean)/std
    return ds
df=standardize_dataset(df,target)
# print df
```

```python
def createMatrix(df,target):
    if target in df.columns:
        df=df.drop(columns=target)
#     df.insert(0,'default',1)
    df=df.values
    return df
train_matrix=createMatrix(df,target)

# train_matrix=createMatrix(train_dataset,target)
# print train_matrix
```

```python
def miniBatches(X,batch_size,train_rows):
      #generate a random list of indices from 0 to train_size without replacement
#         r_idx = np.random.choice(train_rows, train_rows, replace=False)
#         #returns a list of shuffled nums from 0 to train_rows
#         shuffle_input = X[r_idx] #shuffles the input array
#         shuffle_output = output[r_idx]
        mini_batches = []
        for i in range (0, train_rows, batch_size):
            x = X[i:i + batch_size, :]
#             y = shuffle_output[i:i + batch_size, :]
            mini_batches.append([x])
        return mini_batches
```

```python
class NeuralNetwork:
    def __init__(self,cols,targets_num,hidden_size_list,lr,no_hidden_layers):#cols=29
        self.inputLayerSize=cols
        self.outputLayerSize=cols
        self.hiddenLayerSize=hidden_size_list
        self.weights=[]
        self.bias=[]
        self.nhl=no_hidden_layers
        self.weights.append( np.random.randn(self.inputLayerSize,hidden_size_list[0])   * math.sqrt(2.0/self.inputLayerSize))
        self.bias.append(np.random.randn(hidden_size_list[0]))
        for i in range(1,no_hidden_layers):
            self.weights.append( np.random.randn(hidden_size_list[i-1],hidden_size_list[i])   * math.sqrt(2.0/hidden_size_list[i-1]))
            self.bias.append(np.random.randn(hidden_size_list[i]))
        self.weights.append(np.random.randn(hidden_size_list[no_hidden_layers-1],self.outputLayerSize)  * math.sqrt(2.0/hidden_size_list[no_hidden_layers-1]) )
        self.bias.append(np.random.randn(self.outputLayerSize))

        self.learning_rate=lr
#         for i in self.weights:
#             print i.shape


    def saveWeights(self):
        with open("weights.txt","wb") as file:
            pickle.dump(self.weights,file)
        with open("biases.txt","wb") as file:
            pickle.dump(self.bias,file)

    
    def loadWeights(self):
        with open("weights.txt","rb") as file:
            self.weights=pickle.load(file)
        with open("biases.txt","rb") as file:
            self.bias=pickle.load(file)

    
    def forward(self,X,func):
        self.z2=[]
        self.a2=[]
        reduce_dim=len(self.hiddenLayerSize)//2
        
        temp_z2=np.dot(X, self.weights[0])+self.bias[0]
        self.a2.append(X)
        self.z2.append(temp_z2)
        for i in range(self.nhl):
#         print "z2",self.z2
            if func=="sigmoid":
                temp_a2 = self.sigmoid(temp_z2)
            elif func=="tanh":
                temp_a2 = self.tanH(temp_z2)
            elif func=="linear":
                temp_a2=self.linear(temp_z2)
            else:
                temp_a2 = self.ReLU(temp_z2)
            if i==reduce_dim:
                reduced_temp_df=temp_a2
            self.a2.append(temp_a2)
#             print "i:",i  
            temp_z2=np.dot(temp_a2, self.weights[i+1])+self.bias[i+1]
            self.z2.append(temp_z2)
#         print "self.a3",self.z3  
        if func=="sigmoid":
            yHat = self.sigmoid(temp_z2)
        elif func=="tanh":
            yHat = self.tanH(temp_z2)
        elif func=="linear":
            yHat=self.linear(temp_z2)
        else:
            yHat = self.ReLU(temp_z2)
#         print reduced_temp_df.shape
        return yHat,reduced_temp_df
    
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    
    def sigmoid_prime(self,x):  
        return self.sigmoid(x) *(1-self.sigmoid (x))
    
    def tanH_prime(self,x):
        return 1.0 - self.tanH(x)**2
    
    def tanH(self,z):
        return (2/(1+np.exp(-2*z)))-1
        
    def ReLU(self,x):
        return x * (x > 0)

    def dReLU(self,x):
        return 1. * (x > 0)
    
        
    def softmax(self,a):  
        a=np.array(a,dtype=np.float128)
        expA = np.exp(a)
        return expA / expA.sum(axis=1, keepdims=True)
    
    def linear(self,z):
        return z
    
    def linear_prime(self,z):
        return np.ones(z.shape)
        
    def backward(self,yHat,y,X,func):
        dW_list=[]
        dB_list=[]
        m=y.shape[0]
        ########## Phase 1
        dzo = yHat - y
        if func=="sigmoid":
            dzo = np.multiply(dzo,self.sigmoid_prime(yHat))
        elif func=="tanh":
            dzo =np.multiply(dzo,self.tanH_prime(yHat))
        elif func=="linear":
            dzo=np.multiply(dzo,self.linear_prime(yHat))
        else:
            dzo = np.multiply(dzo,self.dReLU(yHat))
        ########## Phases 2
        for i in range(self.nhl,-1,-1):
            dW=np.dot(dzo.T,self.a2[i])/m
            dB=np.sum(dzo, axis=0)/m
#             dzo = np.multiply(dzo.dot(self.weights[i].T) ,self.sigmoid_prime(self.a2[i]))
            if func=="sigmoid":
                dzo = np.multiply(dzo.dot(self.weights[i].T) ,self.sigmoid_prime(self.a2[i]))
            elif func=="tanh":
                dzo = np.multiply(dzo.dot(self.weights[i].T) ,self.tanH_prime(self.a2[i]))
            elif func=="linear":
                dzo = np.multiply(dzo.dot(self.weights[i].T) ,self.linear_prime(self.a2[i]))
            else:
                dzo = np.multiply(dzo.dot(self.weights[i].T) ,self.dReLU(self.a2[i]))
            dW_list.append(dW.T)
            dB_list.append(dB)
        for i in range(self.nhl,-1,-1):
            self.weights[i]-=self.learning_rate * dW_list[self.nhl-i]
            self.bias[i]-=self.learning_rate * dB_list[self.nhl-i]

    def pred(self,yHat):
        y_pred=[]
        num_Rows=yHat.shape[0]
        for i in range(num_Rows):
            y_pred.append(yHat[i].argmax())
        return y_pred
    
      
    def loss(self,yHat,y):
        loss = mean_squared_error(y,yHat)
        print('Loss function value: ', loss)
#         error_cost.append(loss)
        return loss

        
```

```python
def run(train_matrix,func,hidden_size_list,epochs,lr,nhl,batch_size):
    mini_batches=miniBatches(train_matrix,batch_size,train_rows)    
    error=[]
    ep_list=[]
    nn=NeuralNetwork(cols,target_num,hidden_size_list,lr,nhl)
    for epoch in range(epochs):
        y_pred=np.empty([batch_size, cols])
        y_actual=np.empty([batch_size, cols])
        reduced_df=np.empty([batch_size,14 ])
        flag=0
        for mini in mini_batches:
            input_mini=mini[0]
            output_mini=mini[0]
            temp_output,temp_reduce=nn.forward(input_mini,func)
            nn.backward(temp_output,output_mini,input_mini,func)
#             yHat=nn.forward(input_mini,func)
            if flag==0:
#                 y_pred=yHat
                y_pred=temp_output
                y_actual=output_mini
                reduced_df=temp_reduce
                flag=1
            else:
                y_pred=np.concatenate((y_pred, temp_output), 0)
                y_actual=np.concatenate((y_actual, output_mini), 0)
                reduced_df=np.concatenate((reduced_df, temp_reduce), 0)

        if epoch%1==0:
            temp_loss=nn.loss(y_pred,y_actual)
            error.append(temp_loss)
            ep_list.append(epoch)
    print (reduced_df.shape)
    print (reduced_df)
    
    return nn,error,ep_list,reduced_df
```

```python
def plot_graph(error,ep_list):
    plt.figure(figsize=(5, 5))
    plt.plot(ep_list,error,label="epochs vs error")
    plt.xlabel("epochs")
    plt.ylabel("error")
```

```python
from sklearn.cluster import KMeans 
```

```python
def cal_kmeans(final_data):
    kmeans = KMeans(n_clusters=5)  
    kmeans.fit(final_data)
    print(kmeans.cluster_centers_)  
    return kmeans.labels_
```

```python
labels_kmeans=cal_kmeans(reduced_final_data_sig)
train_output=df[target]
label_df=pd.DataFrame(labels_kmeans,columns=["cluster"])
train_output=label_df.join(train_output)
```

```python
# %matplotlib inline
def plot_pie(labels,values,cluster):
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
    y=np.array(values)
    percent = 100.0*y/y.sum()
    legends = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(labels, percent)]
    fig, ax = plt.subplots(figsize=(7, 7))
    wedges, labels = ax.pie(y, colors=colors, shadow=True, startangle=90)
    ax.set_title(cluster)
    wedges, legends, dummy =  zip(*sorted(zip(wedges, legends, y),
                                          key=lambda x: x[2],
                                          reverse=True))

    ax.legend(wedges, legends, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=12)
```

```python
def purity(df,target,cluster,clust_no):
#     print "cluster",cluster
    temp_df=df[df['cluster']==clust_no]
    size_df=len(temp_df)
    uniqueValues=temp_df[target].unique()
    count_list=[]
    labels=[]
    for i in uniqueValues:
        labels.append(i)
        count_list.append(len(df[df[target]==i]))
#         temp=(float(len(df[df[target]==i]))/size_df)*100
#         print i,":",temp,'%'
    plot_pie(labels,count_list,cluster)
```

```python
for i in range(0,5):
    purity(train_output,target,"k"+str(i),i)
```

```python
from sklearn.mixture import GaussianMixture as GMM
def cal_GMM(final_data):
    gmm = GMM(n_components=5).fit(final_data)
    labels = gmm.predict(final_data)
#     print len(labels)
    return labels
```
