## Neural Network

```python
def sigmoid(x):
    return 1 / (1 + np.e ** -x)

def delta_sigmoid(x):
    return x * (1 - x)

def softmax(x):
    shiftx = x - np.max(x).reshape(-1,1)
    exps = np.exp(shiftx)
    return exps / np.sum(exps).reshape(-1,1)

def delta_softmax(x):
    shiftx = x - np.max(x)
    exps = np.exp(shiftx)
    dO = np.zeros(x.shape)
    dO = exps * (np.sum(exps) - exps)
    return dO/(np.sum(exps) ** 2)

def cross_entropy(p,y):
    m = y.shape[0]
#     p = softmax(X)
    y = np.argmax(y)
    log_likelihood = -np.log(p[range(m),y])
    loss = np.sum(log_likelihood) / m
    return loss

def delta_cross_entropy(grad, y):
#     temp = y * (1 / (o_out + eps)) + (1 - y) * (1 / (1 - o_out + eps))
#     temp = -temp
#     return temp
    m = y.shape[0]
#     grad = softmax(X)
    y = np.argmax(y)
    grad[range(m),y] -= 1
    grad = grad/m
    return grad

def delta_mean_squared_error(y, y_):
    return 2 * (y - y_)
```

```python
class NeuralNetwork:
    def __init__(self, learning_rate=0.001):
#         self.X = X
        self.W1 = np.random.rand(784, 10)
        self.W2 = np.random.rand(10, 10)
        self.W3 = np.random.rand(10, 10)
#         self.y = y
        self.y_ = np.zeros(y.shape)
        self.learning_rate = learning_rate
        
    def feedforward(self, X):
        self.h1 = sigmoid(np.dot(X, self.W1))
        self.h2 = sigmoid(np.dot(self.h1, self.W2))
        self.y_ = softmax(np.dot(self.h2, self.W3))
        
    def print_dimensions(self):
      print(self.h1.shape)
      print(self.h2.shape)
      print(self.y_.shape)
      
        
    def backprop(self, X, y):
        loss = delta_mean_squared_error(self.y_, y)
        d_W3 = np.dot(self.h2.T, loss * delta_softmax(self.y_))
        d_W2 = np.dot(self.h1.T, np.dot(loss * delta_softmax(self.y_), self.W3.T) * delta_sigmoid(self.h2))
        d_W1 = np.dot(X.T, np.dot(loss * delta_softmax(self.y_), self.W2.T) * delta_sigmoid(self.h1))
        
        self.W1 += self.learning_rate * d_W1
        self.W2 += self.learning_rate * d_W2
        self.W3 += self.learning_rate * d_W3
        
    def train(self, X, y):
        for i in range(500):
          print(i)
          cost = 0
          for j in range(0,48000,500):
            X = data_train.iloc[j:j+500,1:]
            y = data_train.iloc[j:j+500,0]
            y_mat = np.zeros((X.shape[0], 10))
            for i in range(len(y)):
                y_mat[i][y[i+j]] = 1

            y = y_mat
            self.feedforward(X)
#             print("1: ",self.y_.shape)
#             print("2. ",y.shape)
            self.backprop(X,y)
            cost += cross_entropy(model.y_, y)
#             print("cost_here: ", cost)
#           print(cost)
          print("*******************************************************")
       
    def test(self, X, y):
        h1_test = sigmoid(np.dot(X, self.W1))
        h2_test = sigmoid(np.dot(h1_test, self.W2))
        y_pred = softmax(np.dot(h2_test, self.W3))
#         print(y_pred)
        return y_pred
  
```
